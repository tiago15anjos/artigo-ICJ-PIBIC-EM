# Notebook 02 ‚Äî Limpeza & Padroniza√ß√£o
Objetivo: transformar o `.xlsx` bruto em um `.csv` limpo e recortado (2010-2023),
mantendo a estrutura necess√°ria para as an√°lises do artigo.


```python
import pandas as pd, pathlib

START_YEAR, END_YEAR = 2010, 2023
RAW_FILE = pathlib.Path('data/raw/ICJ_Instituicao_Ano_bolsistas_2010_2023.xlsx')

assert RAW_FILE.exists(), "Arquivo bruto n√£o encontrado. Verifique o caminho."

df = pd.read_excel(RAW_FILE)
print(df.shape)
df.head()
```

```python
# Conferir nomes exatos das colunas
print(df.columns.tolist())
df.dtypes


```

```python
# Renomeia apenas as colunas que usaremos
df = df.rename(columns={
    'Ano': 'ano',
    'Sigla_UF': 'uf',
    'Institui√ß√£o': 'instituicao',
    'Programa': 'programa',
    'Modalidade': 'modalidade',
    'Qtd_Bolsistas_Inst': 'n_bolsistas',
    'Valor_Rs_sum': 'valor_rs'
})

# Conferir se deu certo
df[['ano', 'uf', 'instituicao', 'n_bolsistas', 'valor_rs']].head()
```

```python
# 1. Filtrar per√≠odo (2010 ‚Äì 2024, conforme sua ideia de ampliar)
START_YEAR, END_YEAR = 2010, 2024
df = df.query('ano >= @START_YEAR and ano <= @END_YEAR')

# 2. Manter apenas as colunas essenciais para a an√°lise
df = df[['ano', 'uf', 'instituicao', 'programa',
         'modalidade', 'n_bolsistas', 'valor_rs']]

# 3. Remover registros repetidos (caso existam)
df = df.drop_duplicates()

# Checagem r√°pida
print(df.shape)
df.head()
```

```python
# 1. Garantir tipos num√©ricos
df['ano']         = df['ano'].astype(int)
df['n_bolsistas'] = df['n_bolsistas'].astype(int)
df['valor_rs']    = df['valor_rs'].astype(float)

# 2. Mapear UF ‚Üí Regi√£o
REG_MAP = {
    'AC':'Norte','AL':'Nordeste','AP':'Norte','AM':'Norte','BA':'Nordeste',
    'CE':'Nordeste','DF':'Centro-Oeste','ES':'Sudeste','GO':'Centro-Oeste',
    'MA':'Nordeste','MT':'Centro-Oeste','MS':'Centro-Oeste','MG':'Sudeste',
    'PA':'Norte','PB':'Nordeste','PR':'Sul','PE':'Nordeste','PI':'Nordeste',
    'RJ':'Sudeste','RN':'Nordeste','RS':'Sul','RO':'Norte','RR':'Norte',
    'SC':'Sul','SP':'Sudeste','SE':'Nordeste','TO':'Norte'
}
df['regiao'] = df['uf'].map(REG_MAP)

# Conferir resultado
df[['ano', 'uf', 'regiao', 'n_bolsistas', 'valor_rs']].head()
```

```python
import pathlib

# 1. Criar pasta processed, se ainda n√£o existir
out_path = pathlib.Path('data/processed')
out_path.mkdir(parents=True, exist_ok=True)

# 2. Salvar o CSV
csv_file = out_path / 'icj_clean.csv'
df.to_csv(csv_file, index=False)
print(f"‚úÖ CSV salvo em {csv_file} ‚Äî linhas:", len(df))

# 3. Checagem r√°pida: quantos bolsistas por regi√£o
print("\nBolsistas por regi√£o (2010-2024):")
display(df.groupby('regiao')['n_bolsistas'].sum())
```
